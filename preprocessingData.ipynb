```python
#22022542
import facebook_scraper as fs
#h√†m l·∫•y full comment tham kh·∫£o t·∫°i 'https://github.com/kevinzg/facebook-scraper'
def get_full_comments(post_url):
    POST_ID = post_url

    # number of comments to download -- set this to True to download all comments
    MAX_COMMENTS = 100

    # get the post (this gives a generator)
    gen = fs.get_posts(
       post_urls=[POST_ID],
       options={"comments": MAX_COMMENTS, "progress": True}
    )

    # take 1st element of the generator which is the post we requested
    post = next(gen)

    # extract the comments part
    comments = post['comments_full']
    return str(comments)
```

```python
from facebook_scraper import get_posts
post = next(get_posts(post_urls=["https://facebook.com/tamlyhoctoipham/posts/317149907616740"], options={"comments": True}))
print(post['comments_full'])
```

```python
#22022542
#crawl c√°c post thi·∫øu comment ·ªü tr√™n
#ƒë∆∞a c√°c link b√†i post v√†o 1 list
res = df_null_comment['post_url'].tolist()
#crawl full comment t·ª´ng b√†i d√πng h√†m get_full_comments
for url in res:
    com = get_full_comments(url)
    id = df.loc[(df['post_url'] == url)].index.to_numpy()
    df.loc[id,'comments_full'] = [com]
    print(df.loc[id,'comments_full'])
    
```
